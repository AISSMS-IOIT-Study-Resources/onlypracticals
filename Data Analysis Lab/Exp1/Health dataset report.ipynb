## Pandas Profiling (Profile Report)

```python
import pandas as pd
```

```python
from ydata_profiling import ProfileReport
```

```python
hlt=pd.read_csv("healthexp.csv")
hlt
```

```python
design_report = ProfileReport(hlt, correlations={"auto": {"calculate": False}})
design_report.to_file(output_file='health-report.html')
```

## Sweetviz implementation

```python
import sweetviz as sv
sweet_report = sv.analyze(hlt)
sweet_report.show_html('sweet_health-report.html')
```

```python
df1 = sv.compare(hlt[130:], hlt[:130])
df1.show_html('Compare.html')
```

## Autoviz implementation

```python
from autoviz.AutoViz_Class import AutoViz_Class
AVC = AutoViz_Class()
```

```python
%matplotlib inline
filename = 'healthexp.csv'
sep = ','
dft = AVC.AutoViz(filename, sep, depVar="", dfte=None, header=0, verbose=0,lowess=False,chart_format='svg',max_rows_analyzed=150000,max_cols_analyzed=30)
```
