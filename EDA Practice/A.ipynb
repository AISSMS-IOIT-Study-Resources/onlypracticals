```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filter("ignore")
```

```python
# Load CSV into DataFrame
df = pd.read_csv(r"D:\AISSMS IOIT - AI&DS (628299510)\SE_2023-2024\even_sem\ESE - Target 10 sgpa\EDA 06-06-2024\f.csv")
df.head()
```

```python
# Function to calculate Z-scores
def calculate_z_scores(df):
    z_scores = (df - df.mean()) / df.std()
    return z_scores
```

```python
# Calculate Z-scores
z_scores = calculate_z_scores(df.select_dtypes(include=[np.number]))  # Only numeric columns
```

```python
print("Original DataFrame:")
print(df)
print("\nZ-scores DataFrame:")
print(z_scores)
```

```python
# Data Visualization
# Pairplot to visualize relationships and distributions
sns.pairplot(df.select_dtypes(include=[np.number]))
plt.suptitle("Pairplot of Original DataFrame", y=1.02)
plt.show()
```
